```python
# I like 7533 a lot
from transformers import pipeline
from datasets import load_dataset
import soundfile as sf
import torch
import random

synthesiser = pipeline("text-to-speech", "microsoft/speecht5_tts", device=-1)

embeddings_dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")
speaker_embedding = torch.tensor(embeddings_dataset[7533]["xvector"]).unsqueeze(0)
# You can replace this embedding with your own as well.

speech = synthesiser("Hi! Can I be you kitten? owo", forward_params={"speaker_embeddings": speaker_embedding})

sf.write("speech.wav", speech["audio"], samplerate=speech["sampling_rate"])
```
